This notebook produces the following maps:
* The Arctic continent and the Southern Ocean starting from 60°S = use a Stereographic projections
* The South Atlantic, from 20°S to 50°S. The map should show the following locations: Walvis Bay, Cape Town, Rio de Janeiro, Montevideo

# Output 1: Arctic and Southern Ocean 

```python
#import the necessary libraires 
import matplotlib.pyplot as plt
import matplotlib.path as mpath
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np 
```

```python
#--------------------------------------------------------------------Creating and Plotting the features of the figure--------------------------------------------------------------------
#Creating and Plotting the features of the figure 
#first create the figure 
fig = plt.figure(figsize=(10,10))

#Selected projection for Antarctica = SouthPolarStereo
ax = plt.axes(projection=ccrs.SouthPolarStereo())

#Extent of the figure (map) needs to be set to include the Southern Ocean and Antarctica
#(lon_min, lon_max, lat_min, lat_max)
ax.set_extent([-180, 180, -90, -60], crs=ccrs.PlateCarree())

#Additional features = colours and borders to distinguish between land and sea 
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='aliceblue')
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)
ax.add_feature(cfeature.BORDERS, linestyle=':')

#Gridlines 
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.6, linestyle='--')
gl.top_labels = False
gl.right_labels = False

#--------------------------------------------------------------------Making a circular boundary = this is just for polar aesthetics------------------------------------------------------
#creating 100 evenly spaced angles = defining the points around a circle. 
theta = np.linspace(0, 2*np.pi, 100) 

#computes the x coordinates, ranges between -1 and 1
x = np.sin(theta)

#computes the y coordinates, ranging between -1 and 1
y = np.cos(theta) 

#arranging the coordinates in a single (N,2) array of coordinate pairs
verts = np.column_stack([x, y]) 

#transform circle coordinates, convert points into a Path object = drawable shape
circle = mpath.Path(verts * 0.5 + 0.5)  #*0.5 = shrinks the circle, +0.5 shifts so it is centred in the middle of the axes

#Apply the circular shape as the boundary of the map
ax.set_boundary(circle, transform=ax.transAxes)


#------------------------------------------------------------------------------------Map Title------------------------------------------------------------------------------------------ 
plt.title("The Continent of Antarctic and the surrounding Southern Ocean", fontsize=14, weight='bold')
plt.show()
```

# Output 2: The South Atlantic

```python
#load any additional libraies that were not already loaded for output 1 
from geopy.geocoders import Nominatim 
from geopy.extra.rate_limiter import RateLimiter
```

```python
#--------------------------------------------------------------------Setting up the geolocator and defining the ares of interest-------------------------------------------------------------------
geolocator = Nominatim(user_agent="south_atlantic_map")

#adding a delay to avoid a block by the service 
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

#define the ares of interest to be displayed 
place_names = ["Walvis Bay, Namibia", "Cape Town, South Africa", "Rio de Janeiro, Brazil", "Montevideo, Uruguay"]

#fetch the coordinates 
locations = {}

for place in place_names: 
    loc = geocode(place)
    if loc: 
        locations[place] = (loc.latitude, loc.longitude)

#test, print to verify 
print(locations)

#--------------------------------------------------------------------Creating and Plotting the features of the figure--------------------------------------------------------------------
#Creating and Plotting the features of the figure
#create the figure 
fig = plt.figure(figsize=(12, 8))

#suitable projection
ax = plt.axes(projection=ccrs.PlateCarree())

#setting the extent for the South Atlantic, must include 20-50 S
ax.set_extent([-70, 30, -50, -20], crs=ccrs.PlateCarree())

#Additional features = colours and borders to distinguish between land and sea 
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='aliceblue')
ax.add_feature(cfeature.COASTLINE, linewidth=0.8)

#Gridlines 
gl = ax.gridlines(draw_labels=True, linewidth=0.5, color='gray', alpha=0.6, linestyle='--')
gl.top_labels = False
gl.right_labels = False


#------------------------------------------------------------------------------------------------Plot the locations--------------------------------------------------------------------------------
for name, (lat, lon) in locations.items():
    ax.plot(lon, lat, marker='o', color='red', markersize=6, transform=ccrs.PlateCarree())
    ax.text(lon - 7, lat - 3, name, fontsize=10, transform=ccrs.PlateCarree(), bbox=dict(facecolor='white', alpha=0.6, edgecolor='none'))


#------------------------------------------------------------------------------------Map Title-----------------------------------------------------------------------------------------------------
plt.title("South Atlantic Ocean", fontsize=14, weight='bold')
plt.show()
```
