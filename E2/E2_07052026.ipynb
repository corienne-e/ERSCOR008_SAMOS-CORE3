```python
#--------------------------------FUNCTION: CONVERT ddmm TO DECIMAL DEGREES---------------------------------------------------------
def ddmm2dd(ddmm):
    """
    Converts a position input from degrees and minutes
    to decimal degrees.

    Input:  ddmm.cccc
    Output: dd.cccc
    """

    thedeg = np.floor(ddmm / 100.)
    themin = (ddmm - thedeg * 100.) / 60.

    return thedeg + themin
```

# Part 3: Time Series
Research cruises have an automatic scientific data system that records data from various meteorological and oceanographic instruments together with the position of the ship (the data log). These data are stored in a flat file. The file SAA2_WC_2017_metocean_10min_avg.csv found in the GitHub repository contains the data log from the SA Agulhas II winter cruise in the Southern Ocean in 2017.

Create a script to do the following operations

* Load the CSV file in python, using the appropriate column for the time index. Check that the missing values are properly recognized. 
* The ship reached the southernmost location on July 4th. Use the time indexing to select the data from departure to that date included. All plots will be done using this selected set of data. 
* Plot the time series of temperature with the appropriate labels. Save the figure using the 'grayscale' style 
* Plot a histogram of the salinity distribution using bins of 0.5 psu between 30 and 35. 
* Calculate the mean, standard deviation and the interquartile range for temperature and salinity and present the results in a table. 
* Create a scatter plot of wind speed and air temperature, encoding the latitude information in colour. Wind speed is in meters per second and air temperature in degrees Celsius. Save this figure in PNG with a resolution of 300 DPI 

```python
#Import the necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#---------------------------------------------------------1. LOAD THE CSV FILE--------------------------------------------------------
# Replace with your actual filename
file = "SAA2_WC_2017_metocean_10min_avg.csv"

# Load dataset
# parse_dates converts the time column to datetime format
# index_col sets the time column as the DataFrame index
# na_values ensures missing values are recognised correctly

df = pd.read_csv(file, parse_dates=['TIME_SERVER'], index_col='TIME_SERVER', na_values=['NA', 'NaN', '', '9999', '-999'])

# Check dataset information
print(df.info())

# Check missing values
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
#-------------------------------------------2. SELECT DATA FROM DEPARTURE TO JULY 4TH 2017-------------------------------------------
# Select all data up to and including July 4th
cruise_data = df.loc[:'2017-07-04']

# Display selected range
print("\nSelected data:")
print(cruise_data)
```

```python
#-----------------------------------------------------3. TEMPERATURE TIME SERIES------------------------------------------------------
# Use grayscale plotting style
plt.style.use('grayscale')

# Create figure
fig, ax = plt.subplots(figsize=(10, 5))

# Plot temperature time series
ax.plot(cruise_data.index, cruise_data['TSG_TEMP'], linewidth=1)

# Labels and title
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (°C)')
ax.set_title('Temperature Time Series - Southern Ocean Winter Cruise 2017')

# Rotate dates for readability
plt.xticks(rotation=45)

# Tight layout
plt.tight_layout()

# Save figure
plt.savefig('temperature_timeseries_grayscale.png')

# Show plot
plt.show()
```

```python
#-------------------------------------------------------4. HISTOGRAM OF SALINITY------------------------------------------------------
# Create salinity bins from 30 to 35 every 0.5 psu
bins = np.arange(30, 35.5, 0.5)

# Create histogram
fig, ax = plt.subplots(figsize=(8, 5))

ax.hist(cruise_data['TSG_SALINITY'].dropna(),bins=bins)

# Labels and title
ax.set_xlabel('Salinity (PSU)')
ax.set_ylabel('Frequency')
ax.set_title('Salinity Distribution')

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
#------------------------------------------------------5. SUMMARY STATISTICS TABLE----------------------------------------------------
# Calculate statistics
summary_table = pd.DataFrame({

    'Mean': [
        cruise_data['TSG_TEMP'].mean(),
        cruise_data['TSG_SALINITY'].mean()
    ],

    'Standard Deviation': [
        cruise_data['TSG_TEMP'].std(),
        cruise_data['TSG_SALINITY'].std()
    ],

    'Interquartile Range': [

        cruise_data['TSG_TEMP'].quantile(0.75)
        - cruise_data['TSG_TEMP'].quantile(0.25),

        cruise_data['TSG_SALINITY'].quantile(0.75)
        - cruise_data['TSG_SALINITY'].quantile(0.25)
    ]

},

index=['TSG_TEMP', 'TSG_SALINITY'])

# Display table
print("\nSummary Statistics:")
print(summary_table)

# Optional: save table to CSV
summary_table.to_csv('summary_statistics.csv')
```

```python
# -------------------------------6.SCATTER PLOT: WIND SPEED vs AIR TEMPERATURE LATITUDE ENCODED AS COLOUR-----------------------------
# Reset the plotting style so grayscale is not used
plt.style.use('default')

#create the figure
fig, ax = plt.subplots(figsize=(8, 6))

#make it a scatter plot 
scatter = ax.scatter(
    cruise_data['WIND_SPEED_TRUE'],
    cruise_data['AIR_TEMPERATURE'],
    c=cruise_data['LATITUDE'], 
    cmap='viridis' 
)

# Axis labels
ax.set_xlabel('Wind Speed (m/s)')
ax.set_ylabel('Air Temperature (°C)')

# Title
ax.set_title('Wind Speed vs Air Temperature')

# Add colour bar
cbar = plt.colorbar(scatter)
cbar.set_label('Latitude')

# Tight layout
plt.tight_layout()

# Save figure as PNG with 300 DPI
plt.savefig('windspeed_airtemperature_scatter.png', dpi=300)

# Show plot
plt.show()
```
